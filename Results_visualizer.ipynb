```python
def shivam_function(arousal_list,altruistic_list,data_type,column_no1,column_no2):    
    forbidden_i=[34,38,40,42,44]
    for row in data_type.iterrows(): #altruistic
        arousal_list.append(sum(row[1][column_no1])) #moved
        score=0
        averaging = len(column_no2)
        for i in column_no2:
            if i in forbidden_i:
              if txt == 'Definitely Not' :
                  score+=4
              elif txt == 'Probably Not':
                  score+=2              
              elif txt == 'Possibly':
                score+=0
              elif txt == 'Probably':
                  score+=-2
              elif txt == 'Definitely':
                  score+=-4
              continue

            txt = row[1][i]
            if txt == 'Definitely Not' :
                score+=-4
            elif txt == 'Probably Not':
                score+=-2
            elif txt == 'Possibly':
                score+=0
            elif txt == 'Probably':
                score+=2
            elif txt == 'Definitely':
                score+=4
        altruistic_list.append(score/averaging)
    return arousal_list, altruistic_list
academic = [19,28,29,30,31,33,38,48]
community = [20,23,24,36,37,39,40,42,43,44,45,46]
fnf = [18,21,22,27,35]
general = [25,26,32,34,41,47]
thought = [19,22,24,26,27,31,35,40,41,43,44,45,46,47,48]
action = [18,20,21,23,25,28,29,30,32,33,34,36,37,38,39,42]
```
