```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

metadata_dir = './uplifting.csv'
metadata = pd.read_csv(metadata_dir)

metadata_dir1 = './sad.csv'
metadata1 = pd.read_csv(metadata_dir1)

metadata_dir2 = './neutral.csv'
metadata2 = pd.read_csv(metadata_dir2)
```

```python
# This is for our custom built questionnaire (16 questions)

# Uplifting
final_list = []
for i in range(18,33):
    txt = metadata.columns[i]
    ax1 = metadata[txt]
    size_temp = ax1.size
    score = 0
    for i in range(size_temp):
        if ax1[i] == 'Definitely Not':
            score+=-4
        elif ax1[i] == 'Probably Not':
            score+=-2
        elif ax1[i] == 'Possibly':
            score+=0
        elif ax1[i] == 'Probably':
            score+=2
        elif ax1[i] == 'Definitely':
            score+=4
    final_list.append(score)

    
# Neutral
final_list2 = []
for i in range(18,33):
    txt = metadata2.columns[i]
    ax1 = metadata2[txt]
    size_temp = ax1.size
    score = 0
    for i in range(size_temp):
        if ax1[i] == 'Definitely Not':
            score+=-4
        elif ax1[i] == 'Probably Not':
            score+=-2
        elif ax1[i] == 'Possibly':
            score+=0
        elif ax1[i] == 'Probably':
            score+=2
        elif ax1[i] == 'Definitely':
            score+=4
    final_list2.append(score)
    
    
# Sad    
final_list1 = []
for i in range(18,33):
    txt = metadata1.columns[i]
    ax1 = metadata1[txt]
    size_temp = ax1.size
    score = 0
    for i in range(size_temp):
        if ax1[i] == 'Definitely Not':
            score+=-4
        elif ax1[i] == 'Probably Not':
            score+=-2
        elif ax1[i] == 'Possibly':
            score+=0
        elif ax1[i] == 'Probably':
            score+=2
        elif ax1[i] == 'Definitely':
            score+=4
    final_list1.append(score)
    
```

```python
fig = plt.figure(figsize=(6,5), dpi=200)
left, bottom, width, height = 0.1, 0.3, 2, 1.3
ax = fig.add_axes([left, bottom, width, height]) 
 
width = 0.3
ticks = np.arange(len(final_list))    

ax.bar(ticks, final_list, width, label='Uplifting')

ax.bar(ticks + 2*width, final_list2, width, align="center",
    label='Neutral')

ax.bar(ticks + width, final_list1, width, align="center",
    label='Sad')


ax.set_ylabel('Rating')
ax.set_title('Rating on custom questionnaire')
ax.set_xticks(ticks + width/2)
ax.set_xticklabels(['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9','Q10','Q11','Q12','Q13','Q14','Q15'])

ax.legend(loc='best')
plt.show()
```
