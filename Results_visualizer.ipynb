```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
metadata_dir = '/content/uplifting.csv'
metadata = pd.read_csv(metadata_dir)

metadata_dir1 = '/content/sad.csv'
metadata1 = pd.read_csv(metadata_dir1)

metadata_dir2 = '/content/neutral.csv'
metadata2 = pd.read_csv(metadata_dir2)
```

#Custom Questionnaire Results

```python
# This is for our custom built questionnaire (16 questions)

# Uplifting
final_list = []
for i in range(18,33):
    txt = metadata.columns[i]
    ax1 = metadata[txt]
    size_temp = ax1.size
    score = 0
    for ii in range(size_temp):
        if ax1[ii] == 'Definitely Not':
            score+=-4
        elif ax1[ii] == 'Probably Not':
            score+=-2
        elif ax1[ii] == 'Possibly':
            score+=0
        elif ax1[ii] == 'Probably':
            score+=2
        elif ax1[ii] == 'Definitely':
            score+=4
    final_list.append(score)

    
# Neutral
final_list2 = []
for i in range(18,33):
    txt = metadata2.columns[i]
    ax1 = metadata2[txt]
    size_temp = ax1.size
    score = 0
    for ii in range(size_temp):
        if ax1[ii] == 'Definitely Not':
            score+=-4
        elif ax1[ii] == 'Probably Not':
            score+=-2
        elif ax1[ii] == 'Possibly':
            score+=0
        elif ax1[ii] == 'Probably':
            score+=2
        elif ax1[ii] == 'Definitely':
            score+=4
    final_list2.append(score)
    
    
# Sad    
final_list1 = []
for i in range(18,33):
    txt = metadata1.columns[i]
    ax1 = metadata1[txt]
    size_temp = ax1.size
    score = 0
    for ii in range(size_temp):
        if ax1[ii] == 'Definitely Not':
            score+=-4
        elif ax1[ii] == 'Probably Not':
            score+=-2
        elif ax1[ii] == 'Possibly':
            score+=0
        elif ax1[ii] == 'Probably':
            score+=2
        elif ax1[ii] == 'Definitely':
            score+=4
    final_list1.append(score)
    
```

```python
fig = plt.figure(figsize=(6,5), dpi=200)
left, bottom, width, height = 0.1, 0.3, 2, 1.3
ax = fig.add_axes([left, bottom, width, height]) 
 
width = 0.3
ticks = np.arange(len(final_list))    

ax.bar(ticks, final_list, width, label='Uplifting')

ax.bar(ticks + 2*width, final_list2, width, align="center",
    label='Neutral')

ax.bar(ticks + width, final_list1, width, align="center",
    label='Sad')


ax.set_ylabel('Rating')
ax.set_title('Rating on custom questionnaire')
ax.set_xticks(ticks + width/2)
ax.set_xticklabels(['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9','Q10','Q11','Q12','Q13','Q14','Q15'])

ax.legend(loc='best')
plt.show()
```

```python
def shivam_function(arousal_list,altruistic_list,data_type,column_no1,column_no2):    
    forbidden_i=[34,38,40,42,44]
    for row in data_type.iterrows(): #altruistic
        arousal_list.append(sum(row[1][column_no1])) #moved
        score=0
        for i in column_no2:
            if i in forbidden_i:
              if txt == 'Definitely Not' :
                  score+=4
              elif txt == 'Probably Not':
                  score+=2              
              elif txt == 'Possibly':
                score+=0
              elif txt == 'Probably':
                  score+=-2
              elif txt == 'Definitely':
                  score+=-4
              continue

            txt = row[1][i]
            if txt == 'Definitely Not' :
                score+=-4
            elif txt == 'Probably Not':
                score+=-2
            elif txt == 'Possibly':
                score+=0
            elif txt == 'Probably':
                score+=2
            elif txt == 'Definitely':
                score+=4
        altruistic_list.append(score)
    return arousal_list, altruistic_list
```

#Measures of Self Rating vs custom set of questions

```python
#OUR CUSTOM QUESTIONS
arousal_list = []
altruistic_list=[]
arousal_list,altruistic_list = shivam_function(arousal_list,altruistic_list,metadata,range(6,15),range(18,33)) #happy
arousal_list,altruistic_list = shivam_function(arousal_list,altruistic_list,metadata1,range(6,15),range(18,33)) #sad
arousal_list,altruistic_list = shivam_function(arousal_list,altruistic_list,metadata2,range(6,15),range(18,33)) #neutral

arousal_list = np.asarray(arousal_list)
altruistic_list = np.asarray(altruistic_list)
print(arousal_list)
print(altruistic_list)
r = np.corrcoef(arousal_list, altruistic_list)
print(r[0,1])

fig = plt.figure(figsize=(6,5), dpi=200)
left, bottom, width, height = 0.1, 0.3, 2, 1.3
ax = fig.add_axes([left, bottom, width, height]) 
 
width = 0.3
ticks = np.arange(len(arousal_list))    

y=arousal_list-np.mean(arousal_list)
ax.bar(ticks, y/max(y), width, label='arousal list')

x=altruistic_list-np.mean(altruistic_list)
ax.bar(ticks + width, x/max(x), width, align="center",
    label='altruistic list')


ax.set_ylabel('Rating')
ax.set_title('Rating on General| questionnaire')
ax.set_xticks(ticks + width/2)
ax.set_xticklabels(range(1,50))

ax.legend(loc='best')
plt.show()
```
